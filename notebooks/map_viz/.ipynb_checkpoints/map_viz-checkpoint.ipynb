```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("../../data/Dataset3/data1.csv")
```

```python
df
```

```python
countries = pd.read_csv("../../data/Dataset3/country.csv")
```

```python
countries
```

```python
df = df.merge(countries, on="country_id")
```

## Bwin

```python
dfbw = df[["id", "country_name", "league_id", "season", "home_team_goal", "away_team_goal", "BWH", "BWD", "BWA"]]
```

```python
dfbw.head()
```

```python
dfbw.isna().sum()
```

```python
len(dfbw)
```

```python
dfbw.isna().sum() / len(dfbw)
```

```python
def print_missing(row):
    if pd.isna(row["BWH"]):
        return row["country_name"]
```

```python
missing_countries = dfbw.apply(print_missing, axis=1)
```

```python
missing_countries.dropna().value_counts()
```

```python
dfbw["country_name"].value_counts()
```

```python
dfbw = dfbw.dropna(subset=["BWH"])
```

```python
dfbw.head()
```

```python
def determine_final_result(row):
    hg = int(row["home_team_goal"])
    ag = int(row["away_team_goal"])
    if hg > ag:
        row["result"] = "1"
    elif hg == ag:
        row["result"] = "x"
    else:
        row["result"] = "2"
    return row
```

```python
dfbw = dfbw.apply(determine_final_result, axis = 1)
```

```python
dfbw.head()
```

```python
def determine_bookie_prediction(row):
    h = row["BWH"]
    d = row["BWD"]
    a = row["BWA"]
    
    if h == min([h, d, a]):
        row["prediction"] = "1"
        row["prediction_performance"] = h
    elif d == min([h, d, a]):
        row["prediction"] = "x"
        row["prediction_performance"] = d
    else:
        row["prediction"] = "2"
        row["prediction_performance"] = a
    return row
```

```python
dfbw = dfbw.apply(determine_bookie_prediction, axis = 1)
```

```python
def determine_correct_prediction(row):
    row["correct"] = (row["result"] == row["prediction"])
    return row
```

```python
dfbw = dfbw.apply(determine_correct_prediction, axis = 1)
```

```python
dfbw.head()
```

```python
groupedbw = dfbw.groupby(["country_name", "season"])
```

```python
groupedbw.size()
```

```python
groupedbw.get_group(("Belgium", '2008/2009'))
```

```python
test = groupedbw.get_group(("Belgium", '2008/2009'))
```

## See percentage of accuracy per country

```python
test.head()
```

```python
accuracies = []
for name, group in groupedbw:
    total_acc = group["correct"].sum() / len(group["correct"]) * 100
    #print("Group: ", name, " accuracy: ", total_acc)
    accuracies.append((name[0], name[1], total_acc))
```

```python
#accuracies
```

```python
test = pd.DataFrame(accuracies)
```

```python
test.columns = ["country_name", "season", "accuracy_percentage"]
```

```python
by_season = test.groupby("season")
```

```python
by_season.get_group("2015/2016")
```

```python
for name, group in by_season:
    group.to_csv("bwin/{}.csv".format(name.replace("/","_")))
```
