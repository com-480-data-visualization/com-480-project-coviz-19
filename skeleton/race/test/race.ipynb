```python
df = df[['Div','HomeTeam','Date','year','points']]
df['Month'] =  df.Date.apply(lambda x : x[3:5])
df['Day']=df.Date.apply(lambda x : x[0:2])
df['test']= df['year'].map(str)+'-' + df['Month'].map(str) + '-' + df['Day'].map(str)
df['Date']= df['test']
df = df.drop(columns= ['Day','Month'])
```

```python
def split_year(days):
    number = int(days/36.5)
    return number
```

```python
df['days']= pd.to_datetime(df.Date) - pd.to_datetime(df.year.map(str) +'-01-01') 
df['days']= df['days'].dt.days
df['split']=df.days.apply(split_year)
df['year']= df.year.map(str)+ '.'+ df.split.map(str)
df
```

```python
df.year = df.year.map(float)
df= df.drop(columns=['Date','test','days','split'])
```

```python
test = df.groupby(['Div','HomeTeam','year']).sum().sort_values(by=['Div','HomeTeam'],ascending=False)['points']
```

```python
test = pd.read_csv('test.csv')
maximum_year= test['year'].unique().max()
minimum_year= test['year'].unique().min()
vector = np.arange(minimum_year,maximum_year+0.1,0.1).round(1)
```

```python
teams= test.HomeTeam.unique()
test['before'] = 0
for team in teams:
    df_temp = test[test['HomeTeam']== team]
    years = df_temp.year
    indexes = list(df_temp.index.values)
    count= 0
    for i in indexes:
        if i != min(indexes):
            df_temp.loc[i,'before']= count
        count += df_temp.loc[i].points
        if i != min(indexes):
            df_temp.loc[i,'points']= count
            
    df_temp.loc[min(indexes),'before']= 0
    if team == teams[0]:
        df_try=df_temp
    test[test['HomeTeam']== team]= df_temp
    
for team in teams:
    for i in vector:
        i = i.round(1)
        df_year = pd.DataFrame(columns=['Div','HomeTeam','year','points','before'])
        point = 0
        df_temp = test[test['HomeTeam']== team]
        min_year= df_temp['year'].unique().min()
        indexes = list(df_temp.index.values)
        if i not in df_temp.year.unique():
            if min_year > minimum_year:
                year_comp = np.arange(minimum_year,min_year,0.1).round(1)
                for year in year_comp:
                    data = {'Div': df_temp.loc[indexes[0],'Div'], 'HomeTeam': team, 'year':year.round(1),'points':point,'before':point}
                    df_year=df_year.append(data,ignore_index=True)
            else:
                point = df_temp[df_temp['year']== (i-0.1).round(1)].points.values[-1]
                bef = df_temp[df_temp['year']== (i-0.1).round(1)].before.values[-1]
                div= test[test['HomeTeam']=='Villarreal'].reset_index(drop=True).loc[0,'Div']
                data = {'Div': div, 'HomeTeam': team, 'year':i.round(1),'points':point,'before':bef}
                df_year=df_year.append(data,ignore_index=True)
            test = test.append(df_year)
test=test.sort_values(by='year').reset_index(drop=True)
```
